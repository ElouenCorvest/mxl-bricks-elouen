```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mxlpy import Simulator, scan

import mxlbricks.names as n
from mxlbricks import get_poolman2000
```

```python
get_poolman2000().parameters
```

```python
s = Simulator(get_poolman2000())

s.simulate(9 / 60)

variables, fluxes = s.get_result()

fig, ax = plt.subplots()

ax.plot(variables[n.dhap()])

ax.set_xticks(ax.get_xticks(), [int(i * 60) for i in ax.get_xticks()])

ax.set_ylim(0, 0.3)

plt.show()
```

```python

variable, fluxes = scan.steady_state(
    get_poolman2000(),
    to_scan=pd.DataFrame({n.pi_ext(): np.arange(0.02, 1, 0.02)}),
)

res = pd.concat([variable, fluxes], axis=1)
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(15,5))

axs[0].plot(res["rubisco_carboxylase"] * 1000, marker="D", fillstyle="none", label=r"$J_{\mathrm{Assim}}$")
axs[0].plot((res["ex_g1p"] * 6 + res["starch_phosphorylase"] * -6) * 1000, marker="s", fillstyle="none", label=r"$J_{\mathrm{Starch}}$")
axs[0].plot((res["ex_pga"] + res["ex_gap"] + res["ex_dhap"]) * 1000 * 3, marker="^", fillstyle="none", label=r"$J_{\mathrm{TPT}}$")

axs[0].set_ylim(-100, 200)
axs[0].set_xlim(0, 1)

axs[0].legend()

axs[1].plot(res["3PGA"], marker="d", fillstyle="none", label="PGA")
axs[1].plot(res[n.pi()], marker="s", fillstyle="none", label="Pi")
axs[1].plot(res["3PGA"], marker="D", fillstyle="none", label="PGA")

plt.show()
```
