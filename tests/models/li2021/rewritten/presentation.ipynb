```python
from deltapsi_leaf_oldmodel_new_sim import sim_a_gtype
from mxlpy_version import get_li2021
from mxlpy import make_protocol, Simulator
import math
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
import mxlbricks.names as n
from li2021_odeint_marvin import lightsim_20min_5min, get_params
```

```python
res_li = {}

res_li = lightsim_20min_5min(par=500, params=get_params())

res_li.keys()

res_old = sim_a_gtype({}, "WT", 500)

m = get_li2021()

light_100 = make_protocol([(20 * 60, {"PAR": 500}), (5 * 60, {"PAR": 0})])

s = Simulator(m)
s.simulate_protocol(light_100, time_points_per_step=1000)
res, flux = s.get_result()

keys = {
    "QA_red": ["QAm", "QAm"],
    n.pq_red(): ["PQH2", "PQH2"],
    n.ph("_lumen"): ["pHlumen", "pH_lumen"],
    "Dy": ["Dy", "Dy"],
    n.pottassium("_lumen"): ["Klumen", "K_lumen"],
    n.pc_ox(): ["PC_ox", "PC_ox"],
    "P700_ox": ["P700_ox", "P700_ox"],
    n.zx(): ["Zx", "Zx"],
    "singletO2": ["singletO2", "singletO2"],
    n.fd_red(): ["Fd_red", "Fd_red"],
    n.nadph(): ["NADPH", "NADPH"],
    n.chloride("_lumen"): ["Cl_lumen", "Cl_lumen"],
    n.chloride("_stroma"): ["Cl_stroma", "Cl_stroma"],
    "QA": ["QA", "QA"],
    "P700_red": ["P700_red", "P700_red"],
    n.pq_ox(): ["PQ", "PQ"],
    n.pc_red(): ["PC_red", "PC_red"],
    n.fd_ox(): ["Fd_ox", "Fd_ox"],
    n.nadp(): ["NADP", "NADP"],
    n.vx(): ["Vx", "V"],
    # "driving_force_Cl": ["driving_force_Cl", "driving_force_Cl"],
    n.psbs_pr(): ["PsbS_H", "PsbS_H"],
    "NPQ": ["NPQ", "NPQ"],
    "Phi2": ["Phi2", "Phi2"],
    n.h("_lumen"): ["H_lumen", "H_lumen"],
    "pmf": ["pmf", "pmf"],
    # "delta_pH": ["delta_pH", "delta_pH"],
    # "delta_pH_inVolts": ["delta_pH_inVolts", "delta_pH_inVolts"],
}

fig, axs = plt.subplots(ncols=math.ceil(np.sqrt(len(res.columns))), nrows=math.floor(np.sqrt(len(res.columns))), figsize=(20, 15))

for ax, (key, other) in zip(axs.flatten(), keys.items()):
    ax.plot(res.index / 60, res[key])
    old, new = other
    ax.plot(res_li["time_min"], res_li[new], ls="dashed", color="tab:green")
    ax.plot(res_old["time_axis_min"], res_old[old], ls="dotted", color="red")
    ax.set_title(key)
    
fig.legend(
    [
        Line2D([0], [0], color="tab:blue", ls="solid"),
        Line2D([0], [0], color="tab:green", ls="dashed"),
        Line2D([0], [0], color="tab:red", ls="dotted")
    ],
    ["Mxlpy", "Li2021 Old New Sim", "Li2021 New"],
    ncols=3,
    frameon=False,
    loc="upper center",
    prop={"size": 15},
    bbox_to_anchor=(0.5, 1.05)
)
    
plt.tight_layout()

plt.savefig("comparision_li2021_all.svg", dpi=60, facecolor="white", bbox_inches="tight")
```
